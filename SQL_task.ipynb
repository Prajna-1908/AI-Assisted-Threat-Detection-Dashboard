```
# Step 1: Import SQLite library
import sqlite3

# Step 2: Create a database connection
conn = sqlite3.connect("employee.db")
cursor = conn.cursor()

print("Database connected successfully!")

# Step 3: Create Employee table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Department TEXT,
    Salary REAL
)
""")

conn.commit()
print("Employee table created successfully!")

# Step 4: Display the table schema
print("\nEmployee Table Schema:")
cursor.execute("PRAGMA table_info(Employee);")

schema = cursor.fetchall()

print("CID | Column Name | Data Type | Not Null | Default Value | Primary Key")
print("-" * 70)

for column in schema:
    print(column)

# Step 5: Verify the table has been created
cursor.execute("""
SELECT name
FROM sqlite_master
WHERE type='table' AND name='Employee';
""")

table = cursor.fetchone()

if table:
    print("\nVerification Successful: Employee table exists.")
else:
    print("\nVerification Failed: Employee table does not exist.")

# Step 6: Close the connection
conn.close()
print("\nDatabase connection closed.")
```

SQL Aggregate Functions & Grouping


Prompt:Write a Python program using SQLite that demonstrates the following SQL Aggregate Functions and Grouping operations:


*  Display unique departments using DISTINCT

*  Count the total number of employees

*  Calculate the total salary using SUM()

*  Find the highest salary using MAX()

*  Find the lowest salary using MIN()

*  Group employees by department using GROUP BY

*   Filter grouped results using HAVING

```
# Step 1: Import SQLite library
import sqlite3

# Step 2: Connect to the database
conn = sqlite3.connect("employee.db")
cursor = conn.cursor()

print("Database connected successfully!")

# Step 3: Create Employee table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Department TEXT,
    Salary REAL
)
""")

# Step 4: Delete existing records (optional)
cursor.execute("DELETE FROM Employee")

# Step 5: Insert sample records
employees = [
    (1, 'Alice', 'HR', 50000),
    (2, 'Bob', 'IT', 60000),
    (3, 'Charlie', 'Finance', 55000),
    (4, 'David', 'IT', 70000),
    (5, 'Eva', 'HR', 48000),
    (6, 'Frank', 'Marketing', 45000),
    (7, 'Grace', 'Finance', 65000),
    (8, 'Henry', 'IT', 75000)
]

cursor.executemany("""
INSERT INTO Employee (EmpID, Name, Department, Salary)
VALUES (?, ?, ?, ?)
""", employees)

conn.commit()

# -------------------------------------------------------
# 1. Display unique departments using DISTINCT
# -------------------------------------------------------
print("\n1. Unique Departments")
cursor.execute("SELECT DISTINCT Department FROM Employee")
for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 2. Count total number of employees
# -------------------------------------------------------
print("\n2. Total Number of Employees")
cursor.execute("SELECT COUNT(*) FROM Employee")
print("Total Employees:", cursor.fetchone()[0])

# -------------------------------------------------------
# 3. Calculate total salary using SUM()
# -------------------------------------------------------
print("\n3. Total Salary")
cursor.execute("SELECT SUM(Salary) FROM Employee")
print("Total Salary:", cursor.fetchone()[0])

# -------------------------------------------------------
# 4. Find highest salary using MAX()
# -------------------------------------------------------
print("\n4. Highest Salary")
cursor.execute("SELECT MAX(Salary) FROM Employee")
print("Highest Salary:", cursor.fetchone()[0])

# -------------------------------------------------------
# 5. Find lowest salary using MIN()
# -------------------------------------------------------
print("\n5. Lowest Salary")
cursor.execute("SELECT MIN(Salary) FROM Employee")
print("Lowest Salary:", cursor.fetchone()[0])

# -------------------------------------------------------
# 6. Group employees by department using GROUP BY
# -------------------------------------------------------
print("\n6. Employees Grouped by Department")
cursor.execute("""
SELECT Department,
       COUNT(*) AS TotalEmployees,
       SUM(Salary) AS TotalSalary,
       AVG(Salary) AS AverageSalary
FROM Employee
GROUP BY Department
""")

for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 7. Filter grouped results using HAVING
# -------------------------------------------------------
print("\n7. Departments with More Than 2 Employees")
cursor.execute("""
SELECT Department,
       COUNT(*) AS TotalEmployees
FROM Employee
GROUP BY Department
HAVING COUNT(*) > 2
""")

for row in cursor.fetchall():
    print(row)

# Step 8: Close the connection
conn.close()
print("\nDatabase connection closed.")
```

SQL JOIN Operations

Prompt:Write a Python program using SQLite that demonstrates the following SQL JOIN operations:


*  Create Employee and Department tables

*  Insert records into both tables

*  Display both tables

*  Perform INNER JOIN

*  Perform LEFT JOIN

*  Perform CROSS JOIN

```
# Step 1: Import SQLite library
import sqlite3

# Step 2: Connect to the database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

print("Database connected successfully!")

# Step 3: Create Department table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Department (
    DeptID INTEGER PRIMARY KEY,
    DeptName TEXT NOT NULL
)
""")

# Step 4: Create Employee table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Salary REAL,
    DeptID INTEGER,
    FOREIGN KEY (DeptID) REFERENCES Department(DeptID)
)
""")

# Step 5: Delete existing records (optional)
cursor.execute("DELETE FROM Employee")
cursor.execute("DELETE FROM Department")

# Step 6: Insert records into Department table
departments = [
    (1, 'HR'),
    (2, 'IT'),
    (3, 'Finance'),
    (4, 'Marketing')
]

cursor.executemany("""
INSERT INTO Department (DeptID, DeptName)
VALUES (?, ?)
""", departments)

# Step 7: Insert records into Employee table
employees = [
    (101, 'Alice', 50000, 1),
    (102, 'Bob', 60000, 2),
    (103, 'Charlie', 55000, 3),
    (104, 'David', 70000, 2),
    (105, 'Eva', 48000, None)   # Employee without department
]

cursor.executemany("""
INSERT INTO Employee (EmpID, Name, Salary, DeptID)
VALUES (?, ?, ?, ?)
""", employees)

conn.commit()

# -------------------------------------------------------
# 1. Display Department table
# -------------------------------------------------------
print("\nDepartment Table")
cursor.execute("SELECT * FROM Department")
for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 2. Display Employee table
# -------------------------------------------------------
print("\nEmployee Table")
cursor.execute("SELECT * FROM Employee")
for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 3. INNER JOIN
# -------------------------------------------------------
print("\nINNER JOIN")
cursor.execute("""
SELECT Employee.EmpID,
       Employee.Name,
       Department.DeptName,
       Employee.Salary
FROM Employee
INNER JOIN Department
ON Employee.DeptID = Department.DeptID
""")

for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 4. LEFT JOIN
# -------------------------------------------------------
print("\nLEFT JOIN")
cursor.execute("""
SELECT Employee.EmpID,
       Employee.Name,
       Department.DeptName,
       Employee.Salary
FROM Employee
LEFT JOIN Department
ON Employee.DeptID = Department.DeptID
""")

for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 5. CROSS JOIN
# -------------------------------------------------------
print("\nCROSS JOIN")
cursor.execute("""
SELECT Employee.Name,
       Department.DeptName
FROM Employee
CROSS JOIN Department
""")

for row in cursor.fetchall():
    print(row)

# Step 8: Close the connection
conn.close()

print("\nDatabase connection closed.")
```

SQL Subqueries & Views

Prompt:Write a Python program using SQLite that demonstrates the following SQL operations:

*  Display the Employee table

*  Find employees whose salary is greater than the average salary using a Subquery

*  Find employees working in the IT department using an IN Subquery

*  Check employee existence using EXISTS

*  Create a SQL View

*  Display records from the View

*  Drop the View

*  Verify that the View has been deleted

```
# Step 1: Import SQLite library
import sqlite3

# Step 2: Connect to the database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

print("Database connected successfully!")

# Step 3: Create Department table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Department (
    DeptID INTEGER PRIMARY KEY,
    DeptName TEXT NOT NULL
)
""")

# Step 4: Create Employee table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Salary REAL,
    DeptID INTEGER,
    FOREIGN KEY (DeptID) REFERENCES Department(DeptID)
)
""")

# Step 5: Delete old records
cursor.execute("DELETE FROM Employee")
cursor.execute("DELETE FROM Department")

# Step 6: Insert records into Department table
departments = [
    (1, 'HR'),
    (2, 'IT'),
    (3, 'Finance'),
    (4, 'Marketing')
]

cursor.executemany("""
INSERT INTO Department (DeptID, DeptName)
VALUES (?, ?)
""", departments)

# Step 7: Insert records into Employee table
employees = [
    (101, 'Alice', 50000, 1),
    (102, 'Bob', 60000, 2),
    (103, 'Charlie', 55000, 3),
    (104, 'David', 70000, 2),
    (105, 'Eva', 48000, 4)
]

cursor.executemany("""
INSERT INTO Employee (EmpID, Name, Salary, DeptID)
VALUES (?, ?, ?, ?)
""", employees)

conn.commit()

# -------------------------------------------------------
# 1. Display Employee table
# -------------------------------------------------------
print("\n1. Employee Table")
cursor.execute("SELECT * FROM Employee")
for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 2. Employees whose salary is greater than average salary
# -------------------------------------------------------
print("\n2. Employees with Salary Greater Than Average")

cursor.execute("""
SELECT *
FROM Employee
WHERE Salary >
      (SELECT AVG(Salary) FROM Employee)
""")

for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 3. Employees working in the IT department (IN Subquery)
# -------------------------------------------------------
print("\n3. Employees Working in IT Department")

cursor.execute("""
SELECT *
FROM Employee
WHERE DeptID IN
(
    SELECT DeptID
    FROM Department
    WHERE DeptName='IT'
)
""")

for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 4. Check employee existence using EXISTS
# -------------------------------------------------------
print("\n4. Check Employee Existence (Salary > 65000)")

cursor.execute("""
SELECT EXISTS
(
    SELECT *
    FROM Employee
    WHERE Salary > 65000
)
""")

result = cursor.fetchone()[0]

if result:
    print("Employee exists with salary greater than 65000.")
else:
    print("No employee found.")

# -------------------------------------------------------
# 5. Create a SQL View
# -------------------------------------------------------
cursor.execute("DROP VIEW IF EXISTS EmployeeView")

cursor.execute("""
CREATE VIEW EmployeeView AS
SELECT Employee.EmpID,
       Employee.Name,
       Department.DeptName,
       Employee.Salary
FROM Employee
INNER JOIN Department
ON Employee.DeptID = Department.DeptID
""")

print("\n5. View 'EmployeeView' created successfully.")

# -------------------------------------------------------
# 6. Display records from the View
# -------------------------------------------------------
print("\n6. Records from EmployeeView")

cursor.execute("SELECT * FROM EmployeeView")

for row in cursor.fetchall():
    print(row)

# -------------------------------------------------------
# 7. Drop the View
# -------------------------------------------------------
cursor.execute("DROP VIEW EmployeeView")

print("\n7. View dropped successfully.")

# -------------------------------------------------------
# 8. Verify that the View has been deleted
# -------------------------------------------------------
cursor.execute("""
SELECT name
FROM sqlite_master
WHERE type='view'
AND name='EmployeeView';
""")

view = cursor.fetchone()

if view:
    print("Verification Failed: View still exists.")
else:
    print("8. Verification Successful: View has been deleted.")

# Close connection
conn.close()

print("\nDatabase connection closed.")
```

End-to-End SQL Project

Prompt:Write a Python program using SQLite that demonstrates a complete SQL workflow by performing the following operations:

*  Create Customer and Orders tables

*  Insert records into both tables

*  Display the tables

*  Update a customer record

*  Delete an order

*  Display unique customer cities

*  Calculate total, average, minimum and maximum order amount

*  Group orders by customer

*  Filter grouped data using HAVING

*  Perform an INNER JOIN

*  Perform a LEFT JOIN

*  Find customers with orders greater than the average amount using a subquery

*  Create and display a View

*  Drop the View

```
# Import SQLite library
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("customer_orders.db")
cursor = conn.cursor()

print("Database connected successfully!")

# --------------------------------------------------
# Create Customer Table
# --------------------------------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Customer(
    CustomerID INTEGER PRIMARY KEY,
    CustomerName TEXT NOT NULL,
    City TEXT
)
""")

# --------------------------------------------------
# Create Orders Table
# --------------------------------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Orders(
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    OrderAmount REAL,
    FOREIGN KEY(CustomerID) REFERENCES Customer(CustomerID)
)
""")

# Clear old records (for repeated execution)
cursor.execute("DELETE FROM Orders")
cursor.execute("DELETE FROM Customer")

# --------------------------------------------------
# Insert records into Customer
# --------------------------------------------------
customers = [
    (1, 'Alice', 'Delhi'),
    (2, 'Bob', 'Mumbai'),
    (3, 'Charlie', 'Delhi'),
    (4, 'David', 'Chennai'),
    (5, 'Eva', 'Kolkata')
]

cursor.executemany(
    "INSERT INTO Customer VALUES (?, ?, ?)",
    customers
)

# --------------------------------------------------
# Insert records into Orders
# --------------------------------------------------
orders = [
    (101, 1, 2500),
    (102, 2, 5000),
    (103, 1, 3500),
    (104, 3, 4200),
    (105, 4, 6500),
    (106, 2, 7000)
]

cursor.executemany(
    "INSERT INTO Orders VALUES (?, ?, ?)",
    orders
)

conn.commit()

# --------------------------------------------------
# Display Customer Table
# --------------------------------------------------
print("\nCustomer Table")
cursor.execute("SELECT * FROM Customer")
for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# Display Orders Table
# --------------------------------------------------
print("\nOrders Table")
cursor.execute("SELECT * FROM Orders")
for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# Update Customer Record
# --------------------------------------------------
cursor.execute("""
UPDATE Customer
SET City='Bangalore'
WHERE CustomerID=4
""")
conn.commit()

print("\nCustomer record updated.")

# --------------------------------------------------
# Delete an Order
# --------------------------------------------------
cursor.execute("""
DELETE FROM Orders
WHERE OrderID=103
""")
conn.commit()

print("Order deleted.")

# --------------------------------------------------
# Display Unique Customer Cities
# --------------------------------------------------
print("\nUnique Customer Cities")
cursor.execute("SELECT DISTINCT City FROM Customer")
for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# Aggregate Functions
# --------------------------------------------------
print("\nAggregate Functions")

cursor.execute("""
SELECT
SUM(OrderAmount),
AVG(OrderAmount),
MIN(OrderAmount),
MAX(OrderAmount)
FROM Orders
""")

result = cursor.fetchone()

print("Total Order Amount :", result[0])
print("Average Order Amount :", result[1])
print("Minimum Order Amount :", result[2])
print("Maximum Order Amount :", result[3])

# --------------------------------------------------
# GROUP BY
# --------------------------------------------------
print("\nOrders Grouped by Customer")

cursor.execute("""
SELECT CustomerID,
COUNT(OrderID),
SUM(OrderAmount)
FROM Orders
GROUP BY CustomerID
""")

for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# HAVING
# --------------------------------------------------
print("\nCustomers having Total Order Amount > 6000")

cursor.execute("""
SELECT CustomerID,
SUM(OrderAmount)
FROM Orders
GROUP BY CustomerID
HAVING SUM(OrderAmount) > 6000
""")

for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# INNER JOIN
# --------------------------------------------------
print("\nINNER JOIN")

cursor.execute("""
SELECT Customer.CustomerName,
Orders.OrderID,
Orders.OrderAmount
FROM Customer
INNER JOIN Orders
ON Customer.CustomerID = Orders.CustomerID
""")

for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# LEFT JOIN
# --------------------------------------------------
print("\nLEFT JOIN")

cursor.execute("""
SELECT Customer.CustomerName,
Orders.OrderID,
Orders.OrderAmount
FROM Customer
LEFT JOIN Orders
ON Customer.CustomerID = Orders.CustomerID
""")

for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# Subquery
# --------------------------------------------------
print("\nCustomers with Orders Greater Than Average")

cursor.execute("""
SELECT CustomerName, OrderAmount
FROM Customer
INNER JOIN Orders
ON Customer.CustomerID = Orders.CustomerID
WHERE OrderAmount >
(
SELECT AVG(OrderAmount)
FROM Orders
)
""")

for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# Create View
# --------------------------------------------------
cursor.execute("DROP VIEW IF EXISTS CustomerOrdersView")

cursor.execute("""
CREATE VIEW CustomerOrdersView AS
SELECT Customer.CustomerName,
Customer.City,
Orders.OrderID,
Orders.OrderAmount
FROM Customer
INNER JOIN Orders
ON Customer.CustomerID = Orders.CustomerID
""")

print("\nView Created Successfully.")

# --------------------------------------------------
# Display View
# --------------------------------------------------
print("\nCustomerOrdersView")

cursor.execute("SELECT * FROM CustomerOrdersView")

for row in cursor.fetchall():
    print(row)

# --------------------------------------------------
# Drop View
# --------------------------------------------------
cursor.execute("DROP VIEW CustomerOrdersView")

print("\nView Dropped Successfully.")

# --------------------------------------------------
# Verify View Deletion
# --------------------------------------------------
cursor.execute("""
SELECT name
FROM sqlite_master
WHERE type='view'
AND name='CustomerOrdersView'
""")

if cursor.fetchone():
    print("View still exists.")
else:
    print("View deleted successfully.")

# Close Connection
conn.close()

print("\nDatabase connection closed.")
```
